#### 1. 외부 파일 읽어오기
- CSV 파일

    - path: 파일의 위치(파일명 포함), URL
    - sep: 텍스트 데이터를 필드별로 구분하는 문자
    - header: 열 이름으로 사용될 행의 번호(기본값은 0), 첫 행부터 데이터인 경우 None
    - index_col: 행 인덱스로 사용할 열의 번호 또는 열 이름
    - names: 열 이름으로 사용할 문자열의 리스트
    - skiprows: 처음 몇 줄을 skip할 것인지 설정
    - parse_dates: 날짜 텍스트를 datetime64로 변환할 것인지 설정(기본값은 False)
    - skip_footer: 마지막 몇 줄을 skip할 것인지 설정(숫자 입력)
    - encoding: 텍스트 인코딩 종류를 지정(예: 'utf-8')

```python
# csv파일 읽기: pandas.read_csv("파일경로/파일이름.csv")

import pandas as pd
file_path = './data/read_csv_sample.csv'

df1 = pd.read_csv(file_path)
display(df1)

df2 = pd.read_csv(file_path, header=None)
display(df2)

df3 = pd.read_csv(file_path, index_col='c0')
display(df3)
```
