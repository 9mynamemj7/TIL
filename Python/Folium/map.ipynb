```python
import pandas as pd
import folium
```

```python
df = pd.read_excel("./0310E460.xlsx", engine="openpyxl", sheet_name="맹견", index_col = "아이디")
df2 = pd.concat( [df.위도, df.경도], axis=1 )
```

```python
df.시설명
```

```python
g_map = folium.Map(location=[37.240, 127.199], zoom_start=15)
g_map
```

```python
for i in range(len(df2)):
    marker = folium.Marker([df2.iloc[i][0], df2.iloc[i,1]],
                                            popup=df.시설명.iloc[i],
                                            icon = folium.Icon(color='blue')).add_to(g_map)
```

```python
g_map
```

```python
folium.Map.save(g_map, "./map.html")
```
