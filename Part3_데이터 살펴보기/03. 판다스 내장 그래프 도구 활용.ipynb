#### 판다스 내장 그래프 도구 활용
- 선 그래프: plot

```python
# DataFrame 객체.plot()

import pandas as pd

df = pd.read_excel("./data/남북한발전전력량.xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South','North']
df_ns.columns = df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경
display(df_ns.head())

print("\n")

# 선 그래프 그리기
df_ns.plot()
```

```python
tdf_ns = df_ns.T
display(tdf_ns)
print('\n')
tdf_ns.plot()
```

- 막대 그래프: bar

```python
# DataFrame 객체.plot(kind='bar')
import pandas as pd

df = pd.read_excel("./data/남북한발전전력량.xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South','North']
df_ns.columns = df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경

tdf_ns = df_ns.T
tdf_ns.plot(kind='bar')
```

- 히스토그램: hist

```python
# DataFrame 객체.plot(kind='hist')
import pandas as pd

df = pd.read_excel("./data/남북한발전전력량.xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South','North']
df_ns.columns = df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경

tdf_ns = df_ns.T
tdf_ns = tdf_ns.astype(int)
tdf_ns.info()
tdf_ns.plot(kind='hist')
```

- 산점도: scatter

```python
# DataFrame 객체.plot(kind='scatter')
import pandas as pd

df = pd.read_csv("./data/auto-mpg.csv")
df.columns = ['mpg', 'cylinder', 'displacement', 'horsepower', 'weight',
            'acceleration', 'model year', 'origin', 'name']

df.plot(x='weight', y='mpg', kind='scatter')
```

- 박스 플롯: box

```python
# DataFrame 객체.plot(kind='box')
import pandas as pd

df = pd.read_csv("./data/auto-mpg.csv")
df.columns = ['mpg', 'cylinder', 'displacement', 'horsepower', 'weight',
            'acceleration', 'model year', 'origin', 'name']

df[['mpg','cylinder']].plot(kind='box')
```
