#### 인덱스 활용
- 특정 열을 행 인덱스로 설정

```python
import pandas as pd
exam_data = {"이름": ['서준', '우현', '인아'],
            "수학": [90, 80, 70],
            "영어": [98, 89 ,95],
            "음악": [85, 95, 100],
            "체육": [100,90,90]}
df = pd.DataFrame(exam_data)
display(df)

ndf = df.set_index(['이름'])
display(ndf)

ndf2 = df.set_index('음악')
display(ndf2)

ndf3 = df.set_index(['수학','음악'])
display(ndf3)
```

```python
# 행 인덱스 재배열
dict_data = {'c0': [1,2,3], 'c1': [4,5,6], 'c2': [7,8,9], 'c3': [10,11,12], 'c4': [13,14,15]}

df = pd.DataFrame(dict_data, index = ['r0', 'r1', 'r2'])
display(df)

# 인덱스를 재지정
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf = df.reindex(new_index)
display(ndf)

ndf2 = df.reindex(new_index, fill_value=0)
display(ndf2)
```

```python
# 행인덱스 초기화
display(df)

ndf = df.reset_index()
display(ndf)
```

```python
# 행인덱스 기준으로 데이터프레임 정렬
ndf = df.sort_index()
display(ndf)
```

- 열기준 정렬

```python
# DataFrame 객체.sort_values()
# ascending은 오름차순인지 내림차순인지 정하는 파라미터
display(df)

ndf = df.sort_values(by='c1',ascending=True)
display(ndf)
ndf2 = df.sort_values(by='c1',ascending=False)
display(ndf2)
```
